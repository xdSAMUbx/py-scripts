```python
from pathlib import Path
from habanero import Crossref, cn
import re2
```

```python
path = r"G:\Otros ordenadores\Mi PC\IGAC\archivos_tesis\bibliografia_relacionada\dois_nombres_bibliografia.txt"
path2 = r"G:\Otros ordenadores\Mi PC\IGAC\archivos_tesis\bibliografia_relacionada\dois.bib"
path = Path(path)

pat = re2.compile(r'^-\s+DOI\:\s+(\S+)')
with open(path, "r", encoding="utf-8") as f, open(path2, "w", encoding="utf-8") as o:
    lines = f.readlines()
    n = len(lines)
    for i, linea in enumerate(lines):
        linea = linea.strip()
        m = pat.match(linea)

        if not m:
            continue

        doi = m.group(1)
        doi = doi.replace("https://doi.org/", "")
        doi = doi.replace("http://doi.org/", "")
        doi = doi.replace("http://dx.doi.org/", "")
        cita = cn.content_negotiation(ids = doi, format = "bibentry")
        o.write(cita + "\n")

    o.close()
```
