```python
from pathlib import Path
import re
from pypdf import PdfWriter, PdfReader
```

```python
path = "G:\\Otros ordenadores\\Mi PC\\Universidad\\9 Semestre\\Monitorias\\Elsy\\70\\act_mineralogia"

pdfs = Path(path)

def extraer_numero(archivo):
    match = re.match(r'^(\d+)_', archivo.name)
    return int(match.group(1)) if match else 0

if pdfs.is_dir():
    writer = PdfWriter()
    
    archivos = sorted(pdfs.glob("*.pdf"), key=extraer_numero)
    
    for archivo in archivos:
        print(f"Agregando: {archivo.name}")  # para verificar el orden
        reader = PdfReader(archivo)
        for page in reader.pages:
            writer.add_page(page)
    
    with open(pdfs / "presentaciones_gr71.pdf", "wb") as output:
        writer.write(output)
    
    print("✅ PDF unido correctamente!")

```
