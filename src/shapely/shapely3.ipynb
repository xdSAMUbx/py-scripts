```python
from pathlib import Path
import geopandas as gpd
import pandas as pd
import numpy as np
from shapely import Point
import matplotlib.pyplot as plt 
```

```python
dataPath = Path.cwd().parent.parent / "data"
path = str(dataPath / "PluviDBFinal.csv")
path1 = str(dataPath /  "Municipios_Cundinamarca.shp")

df = pd.read_csv(path)
```

```python
gdf = gpd.GeoDataFrame(df,geometry=gpd.points_from_xy(df["Longitud"], df["Latitud"]), crs="EPSG:4326")
colImp = ("2020-01","Nombre","Latitud","Longitud","Elevacion","Municipio","Departamento","geometry")
gdf = gdf.drop(columns=[col for col in gdf.columns if not col.startswith(colImp)])
gdf.to_crs(crs="EPSG:3116",inplace=True)
gdf.plot()
```

```python
import os
os.environ['SHAPE_RESTORE_SHX'] = 'YES'
gdf1 = gpd.read_file(path1)
gdf1.set_crs(crs='EPSG:3116', inplace=True)
gdf1.plot()
```
